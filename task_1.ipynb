```python
from pandas.io.stata import excessive_string_length_error


def calculate_roi(profit, investment):
    roi = (profit / investment) * 100
    return roi

campaign_roi = calculate_roi(5000, 20000)
total_roi_bonus = campaign_roi * 1000

print(total_roi_bonus)
```

```python
def calculate_net_profit(revenue, fixed_costs):
    if revenue < fixed_costs:
        return revenue - fixed_costs
    net_profit = (revenue - fixed_costs) - (revenue - fixed_costs)*0.18

    return net_profit

revenue = 500000
fixed_costs = 350000

print(calculate_net_profit(revenue,fixed_costs))
```

```python
def check_credit_score(income, debt):
    if income <= 0:
        return "Refused: Not enough funds"
    credit_load = (debt / income) * 100
    if credit_load > 40:
        return "Refused: High Credit Load"
    else:
        return "Approved"
```
