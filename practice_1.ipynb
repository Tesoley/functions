```python
def calculate_net_profit(revenue, fixed_costs):
    if revenue < fixed_costs:
        return revenue - fixed_costs
    net_profit = (revenue - fixed_costs) - (revenue - fixed_costs)*0.18

    return net_profit

revenue = 500000
fixed_costs = 350000

print(calculate_net_profit(revenue,fixed_costs))
```

```python
def check_credit_score(income, debt):
    if income <= 0:
        return "Refused: Not enough funds"
    credit_load = (debt / income) * 100
    if credit_load > 40:
        return "Refused: High Credit Load"
    else:
        return "Approved"

income = 40000
debt = 18000

print(check_credit_score(income, debt))
```

```python
def calculate_total(*args):
    total_bonus = 0

    for n in args:
        if n < 10000:
            total_bonus += 0
        elif 10000 <= n <= 50000:
            total_bonus += n * 0.05
        elif n > 50000:
            total_bonus += n * 0.1


    return total_bonus

calculate_total(5000, 12000, 60000, 8000)
print(f"Your total is {calculate_total(5000, 12000, 60000, 8000)}")
```
